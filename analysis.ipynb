```python
# Saving the Clean Dataset Variable
# This is the dataframe I will be using in this further project
df_clean = df_encoded.copy()
```

# TASK 3 - Model Building

```python
# 1. Importing required libraries
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.ensemble import RandomForestRegressor
from sklearn.metrics import (
    mean_absolute_error,
    mean_squared_error,
    r2_score
)
```

```python
# 2. Defining Features and Target
X = df_clean.drop("price", axis=1)
y = df_clean["price"]
```

```python
# Verifying Shapes
print("Features Shape:", X.shape)
print("Target Shape:", y.shape)
```

```python
# Train-Test Split (80/20)
X_train, X_test, y_train, y_test = train_test_split(
    X,
    y,
    test_size=0.20,
    random_state=42
)
```

```python
# Verifying Split
print("Training Features:", X_train.shape)
print("Testing Features:", X_test.shape)
print("Training Target:", y_train.shape)
print("Testing Target:", y_test.shape)
```

```python
# Training Linear Regression
lr = LinearRegression()
lr.fit(X_train, y_train)
```

```python
# Making Predictions
lr_pred = lr.predict(X_test)
```

```python
# Evaluate Linear Regression
lr_mae = mean_absolute_error(y_test, lr_pred)

lr_rmse = np.sqrt(mean_squared_error(y_test, lr_pred))

lr_r2 = r2_score(y_test, lr_pred)

print("Linear Regression Results: ")
print("--------------------------")
print("MAE :", lr_mae)
print("RMSE:", lr_rmse)
print("R²  :", lr_r2)
```

```python
# 1. Training Random Forest Regressor
rf = RandomForestRegressor(
    n_estimators=200,
    random_state=42
)

rf.fit(X_train, y_train)
```

```python
# 2. Making Predictions
rf_pred = rf.predict(X_test)
```

```python
# 3. Evaluate Random Forest
rf_mae = mean_absolute_error(y_test, rf_pred)

rf_rmse = np.sqrt(mean_squared_error(y_test, rf_pred))

rf_r2 = r2_score(y_test, rf_pred)

print("Random Forest Results")
print("----------------------")
print("MAE :", rf_mae)
print("RMSE:", rf_rmse)
print("R²  :", rf_r2)
```

```python
# Comparing Both Models
comparison = pd.DataFrame({
    "Model": ["Linear Regression", "Random Forest"],
    "MAE": [lr_mae, rf_mae],
    "RMSE": [lr_rmse, rf_rmse],
    "R2 Score": [lr_r2, rf_r2]
})

comparison
```

## Model Comparison

The Linear Regression model performed better than the Random Forest Regressor on this dataset.

- **MAE (Mean Absolute Error):** ₹970,043
- **RMSE (Root Mean Squared Error):** ₹1,324,507
- **R² Score:** 0.653

The model achieved lower MAE and RMSE values, indicating more accurate predictions, and a higher R² score, showing that it explains a larger portion of the variation in house prices.

Based on these results, Linear Regression was selected as the better-performing model for house price prediction.

# TASK 4 - Visualization

```python
# House Price Distribution using Histogram
import matplotlib.pyplot as plt
import seaborn as sns

plt.figure(figsize=(8,5))

sns.histplot(df["price"], bins=30, kde=True)

plt.title("Distribution of House Prices")
plt.xlabel("House Price")
plt.ylabel("Frequency")

plt.tight_layout()

plt.savefig("histogram.png")

plt.show()
```

The distribution of house prices is positively skewed. Most houses fall within the mid-price range, while a small number of premium properties have significantly higher prices, creating a long right tail.

```python
# Correlation Heatmap
plt.figure(figsize=(12,8))
sns.heatmap(
    df_clean.corr(),
    annot=True,
    fmt=".2f",
    cmap="coolwarm"
)
plt.title("Correlation Heatmap")
plt.tight_layout()
plt.savefig("heatmap.png")
plt.show()
```

The heatmap indicates that area and bathrooms have the strongest positive relationship with house prices. Features such as air conditioning, number of stories, and parking availability also contribute positively to property value. Unfurnished houses tend to have lower prices.

```python
# Actual vs Predicted House Price
plt.figure(figsize=(8,5))

plt.scatter(y_test, lr_pred, alpha=0.7)

plt.plot(
    [y_test.min(), y_test.max()],
    [y_test.min(), y_test.max()],
    'r--',
    linewidth=2
)

plt.xlabel("Actual Price")
plt.ylabel("Predicted Price")
plt.title("Actual vs Predicted House Prices")

plt.tight_layout()

plt.savefig("scatter.png")

plt.show()
```

The Actual vs Predicted scatter plot shows a strong positive relationship between actual and predicted house prices. Although some prediction errors are present, the model captures the overall pricing trend effectively.
